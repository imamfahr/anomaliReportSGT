```python
wb = openpyxl.load_workbook('Weekly Maintenance Performance SGT Week 05 - Anomali (1).xlsx')
sheet = wb['SGT P.Metrics']

data_rows = []
for row in sheet['A7':'G47']:
    data_cols = []
    for cell in row:
        data_cols.append(cell.value)
    data_rows.append(data_cols)

#drop and add first row to be index 
report = pd.DataFrame(data_rows)
report.drop([1,2],axis=1,inplace=True)
report.columns = report.iloc[0]

#first row to be dropped
report = report.drop(0).reset_index(drop=True)
```

```python
#set underperformed unit type to be in a dataframe
anomalies = report.loc[report['Week 04'] <= 0.9]
anomalies.reset_index(drop=True)
```
